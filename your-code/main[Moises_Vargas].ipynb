# Challenge 1 - Analysis of Variance

In this part of the lesson, we will perform an analysis of variance to determine whether the factors in our model create a significant difference in the group means. We will be examining a dataset of FIFA players. We'll start by loading the data using the code in the cell below.

```python
# Run this code:
fifa = pd.read_csv('fifa.csv')
```

Let's examine the dataset by looking at the `head`.

```python
# Your code here:

fifa .head()
```

```python
fifa.info()
```

Player's values are expressed in millions of euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell. Assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions. 

```python
# Your code here:
import re
```

```python
fifa['ValueNumeric'] = fifa['Value'].str.replace('[€,M,K]', '')
fifa['ValueNumeric'] = pd.to_numeric(fifa.ValueNumeric)
fifa['ValueNumeric']
```

#### We'd like to determine whether a player's preffered foot and position have an impact on their value. 

Using the `statsmodels` library, we are able to produce an ANOVA table without munging our data. Create an ANOVA table with value as a function of position and preferred foot. Recall that pivoting is performed by the `C` function.

Hint: For columns that have a space in their name, it is best to refer to the column using the dataframe (For example: for column `A`, we will use `df['A']`).

```python
# Your code here:
fifa.columns = fifa.columns.str.replace(' ', '_')
fifa.columns
formula = 'ValueNumeric ~ C(Position) + C(Preferred_Foot)'
# formula = 'fifa["ValueNumeric"] ~ C(fifa["Position"]) + C(fifa["Preferred_Foot"])'
model = ols(formula=formula, data=fifa).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

What is your conclusion from this ANOVA?

```python
# Your conclusions here:
'''Se rechaza la hipotesis de nula para la ANOVA de dos vías
H0:  μ1=μ2=μ3, al menos una de las variables de pi preferido 
y posición impacta en el valor de cada jugador. 
'''
```

After looking at a model of both preffered foot and position, we decide to create an ANOVA table for nationality. Create an ANOVA table for numeric value as a function of nationality.

```python
# Your code here:
formula = 'ValueNumeric ~ C(Nationality)'
model = ols(formula=formula, data=fifa).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

What is your conclusion from this ANOVA?

```python
'''Se rechaza la hipótesis nula, las medias son difernetes entre el valor del jugador y la nacionalidad '''
```

```python
fifa.shape
fifa_clean = fifa.dropna()
fifa_clean.head()
```

Using the FIFA dataset, in the cell below, create a linear model predicting value using stamina and sprint speed. create the model using `statsmodels`. Print the model summary.

Hint: remember to add an intercept to the model using the `add_constant` function.

```python
# Your code here:
corr = fifa_clean.corr()
corr
```

```python
x = sm.add_constant(fifa_clean[['Stamina', 'SprintSpeed']])#creamos variable x con las columnas correspondinetes a las variables independientes
y = fifa_clean['ValueNumeric']#Asignamos a una variable y la columna corrspondinete a la variable dependiente.
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

```python
# Your conclusions here:
modelo = sm.OLS(y,x).fit()
print(modelo.summary())
'''Con base en los resultados del modelo, el valor p-value para el estadístico t
de la Sprint Speed es mayor al 0.05, por lo que esta no tiene una relación causal con la valor del jugador. Asimismo, 
el R^2 es igual a cero, que corresponde al coeficiente de determinación múltiple, cuanto mayor 
sea la varianza explicada por el modelo de regresión, más cerca estarán los puntos de
los datos de la línea de regresión ajustada, por lo que en este caso las variables no explican el valor de 
los jugadorres.
Además el p-value del estadistico F es mayor a 0.05, por lo que no se rechaza la H0, el modelo no esta bien especificado.
'''
```

Next, create a second regression model predicting value using potential. Create the model using `statsmodels` and print the model summary. Remember to add a constant term.

```python
# Your code here:
x = sm.add_constant(fifa_clean['Potential'])#creamos variable x con las columnas correspondinetes a las variables independientes
y = fifa_clean['ValueNumeric']#Asignamos a una variable y la columna corrspondinete a la variable dependiente.
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

```python
# Your conclusions here:
modelo = sm.OLS(y,x).fit()
print(modelo.summary())
'''
Con base en los resultados del modelo, el valor p-value para el estadístico t es menor al 0.05
para la variable Potential, existe una relación causal con el valor del jugador. Asimismo, el 
R^2 es muy cercano a cero, que corresponde al coeficiente de determinación múltiple, cuanto mayor 
sea la varianza explicada por el modelo de regresión, más cerca estarán los puntos de
los datos de la línea de regresión ajustada, por lo que en este caso la variable potential no explican el valor de 
los jugadorres.
Por otro lado, el p-value del estadistico F es mayor a 0.05, por lo que se rechaza la H0, es decir, existe
relación funcional, sin embargo esto no hace mucho sentido con el valor del R^2.
'''
```
